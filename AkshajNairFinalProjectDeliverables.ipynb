# Question 5a Data Dictionary Continued
1.	track_uri: The Spotify ID for the track
2.	album_name: The album name in which the track appears
3.	track_name: Name of the track
4.	artists: The artists' names who performed the track. If there is more than one artist, they are separated by a ;
5.	Release Date: The date of release for the track
6.	duration_ms: The track length in milliseconds
7.	popularity: The popularity of a track is a value between 0 and 100, with 100 being the most popular. The popularity is calculated by algorithm and is based, in the most part, on the total number of plays the track has had and how recent those plays are. Generally speaking, songs that are being played a lot now will have a higher popularity than songs that were played a lot in the past. Duplicate tracks (e.g. the same track from a single and an album) are rated independently. Artist and album popularity is derived mathematically from track popularity.

8.	explicit: Whether or not the track has explicit lyrics (true = yes it does; false = no it does not OR unknown)

9.	Added By: the username of which one of my friends added the song

10.	Added At: time the songs were added to the playlist
11.	Genres: The genre in which the track belongs
12.	Record Label: The label associated with the publication of the songs
13.	danceability: Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable
14.	energy: Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale
15.	key: The key the track is in. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1
16.	loudness: The overall loudness of a track in decibels (dB)
17.	mode: Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0
18.	speechiness: Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks
19.	acousticness: A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic
20.	instrumentalness: Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content
21.	liveness: Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live
22.	valence: A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry)
23.	tempo: The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration
24.	time_signature: An estimated time signature. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure). The time signature ranges from 3 to 7 indicating time signatures of 3/4, to 7/4.

```python
#Aggregation

pl1_shuffled["Mood"] = pd.cut(
    pl1_shuffled["Valence"],
    bins=[0, 0.3, 0.6, 1.0],
    labels=["Sad", "Neutral", "Happy"]
)

mood_counts = pd.crosstab(
    pl1_shuffled["Artist Name(s)"],
    pl1_shuffled["Mood"]
)

print(mood_counts)
```

```python
#Cross Tabulation

#I have mostly numerical columnbs but my testing might be benefited by having some categories based off of the numerical columns

pl1_shuffled["Tempo Category"] = pd.cut(
    pl1_shuffled["Tempo"],
    bins=[0, 90, 120, 200],
    labels=["Slow", "Medium", "Fast"]
)

tempo_crosstab = pd.crosstab(
    pl1_shuffled["Artist Name(s)"],
    pl1_shuffled["Tempo Category"]
)

print(tempo_crosstab)
```

To get more interpretable insights from my mostly numeric dataset, I created categorical versions of key continuous features. By converting valence into a three-level “Mood” category and tempo into Slow/Medium/Fast buckets, I can run cross-tabulations that reveal how different artists contribute to the emotional tone and pacing of the playlist. These categorized features make patterns much easier to spot compared to raw numerical values—for example, which artists tend to produce “Happy” vs. “Sad” songs or who leans toward faster tempos. Creating these categories is also helpful later when building classification models, since models often learn more effectively from clear, human-interpretable groupings embedded in the data.

```python
#PCA

from sklearn.decomposition import PCA
import numpy as np

# Only numeric features
X_train = train_scaled[numeric_cols]

# Fit PCA
pca = PCA()
pca.fit(X_train)

# Transform the training set
X_train_pca = pca.transform(X_train)

plt.figure(figsize=(8,5))

plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot: PCA Explained Variance')
plt.grid()
plt.show()

for i, var in enumerate(pca.explained_variance_ratio_):
    print(f"PC{i+1}: {var:.3f}")

loadings = pd.DataFrame(pca.components_.T, 
                        columns=[f'PC{i+1}' for i in range(len(numeric_cols))],
                        index=numeric_cols)

print(loadings)

loadings.abs().sort_values(by='PC1', ascending=False).head(10)
```

# Last Steps 

After consulting with Professor Benson, I will run one last step and utilize my PCA results to move forward with my analysis in K-Means Clustering.

Since at the 4-5 mark, almost 80% of the variance can be explained I will set the cuttoff at 4 for the k-means clustering.

```python
#Reducing Data to 4 Components & Choosing number of clusters

n_components = 4

X_pca = X_train_pca[:, :n_components]

#Elbow method to verify

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []

for k in range(2, 10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_pca)
    inertias.append(km.inertia_)

plt.plot(range(2, 10), inertias, marker='o')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.grid()
plt.show()

#Silhouette Score
from sklearn.metrics import silhouette_score

sil_scores = {}

for k in range(2, 10):
    km = KMeans(n_clusters=k, random_state=42)
    labels = km.fit_predict(X_pca)
    sil_scores[k] = silhouette_score(X_pca, labels)

sil_scores
```

```python
#Kmeans - we're going to go with 3 so its easier to interpret even though, 2 was the statistically best choice

from sklearn.cluster import KMeans

# Use first 4 principal components
X_pca = X_train_pca[:, :4]

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_pca)

train_scaled['Cluster'] = clusters

```

```python
#Interpretation

cluster_profile = train_scaled.groupby('Cluster')[numeric_cols].mean()
cluster_profile
```

```python
#Samples for each cluster (statistically closest to cluster parts)

import numpy as np
from sklearn.metrics import pairwise_distances_argmin_min

X_pca = X_train_pca[:, :4]

kmeans = KMeans(n_clusters=3, random_state=42).fit(X_pca)

closest_indices, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, X_pca)

closest_indices

train_scaled.iloc[closest_indices][['Track Name', 'Artist Name(s)', 'Cluster']]
```
